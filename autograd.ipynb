```python
x=torch.tensor(3.0)
y=torch.tensor(0.0)
```

```python
w=torch.tensor(1.0,requires_grad=True)
b=torch.tensor(0.0,requires_grad=True)
```

```python
w,b
```

```python
## forward propogation
z=w*x+b
z
```

```python
y_pred=torch.sigmoid(z)
y_pred
```

```python
import torch.nn.functional as f 

loss=f.binary_cross_entropy(y_pred,y)
```

```python
loss
```

```python
# derivative of loss wrt w and b
loss.backward()
```

```python
print(w.grad,b.grad)
```
