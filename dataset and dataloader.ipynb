```python
X,y=make_classification(
  n_samples=10,
  n_features=2,
  n_informative=2,
  n_redundant=0,
  n_classes=2,
  random_state=42
)
```

as we know in pytorch we have to given our x trai data into float 32 and y train daya in int 64 

```python
X=torch.tensor(X,dtype=torch.float32)
y=torch.tensor(y,dtype=torch.long)
```

```python
from torch.utils.data import Dataset,DataLoader
```

```python
class custom_datasets(Dataset):
  def __init__(self,features,labels):
    self.features=features
    self.labels=labels
  def __len__(self):
    return self.features.shape[0]## no of rows in our dataset 
  def __getitem__(self,index):
    return self.features[index],self.labels[index]
```

```python
dataset=custom_datasets(X,y)
```

```python
len(dataset)
```

```python
## if you want to access any index rows you can do 
dataset[0]
```

```python
dataloader=DataLoader(dataset,batch_size=2,shuffle=True)
```
