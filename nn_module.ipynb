```python
import torch 
import torch.nn as nn 
```

```python
class Model(nn.Module):
  def __init__(self,num_features):

    super().__init__()
    self.Linear=nn.Linear(num_features,1)
    self.sigmoid=nn.Sigmoid()
  def forward(self,features):## this forward function perfrom wx+b 
    out=self.Linear(features)
    out=self.sigmoid(out)
    return out 
```

```python
## create a dataset 
features=torch.rand(10,5)
model=Model(features.shape[1])
model(features)
```

```python
## now here you can also see your weight value and bias valuue that are used 
model.Linear.weight
```

```python
## now lets make a simple neuron network with 5 input features one hidden layer 3 neuron and one output neuron 
class models(nn.Module):
  def __init__(self,num_features):
    super().__init__()
    self.linear1=nn.Linear(num_features,3)
    self.relu=nn.ReLU()
    self.linear2=nn.Linear(3,1)
    self.sigmoid=nn.Sigmoid()
  def forward(self,features):
    out=self.linear1(features)
    out=self.relu(out)
    out=self.linear2(out)
    out=self.sigmoid(out)
    return out 
    
```

```python
features=torch.rand(10,5)
model=models(features.shape[1])
model(features)
```

## now lets make this using sequential 

```python
class Models(nn.Module):
  def __init__(self,num_features):
    super().__init__()
    self.sequential=nn.Sequential(
      nn.Linear(num_features,3),
      nn.ReLU(),
      nn.Linear(3,1),
      nn.Sigmoid()
    )
  def forward(self,features):
    out=self.sequential(features)
    return out
```

```python
features=torch.rand(10,5)
model=Models(5)
model(features)
```
